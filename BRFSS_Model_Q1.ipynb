## Our dataset is imbalanced.
As almost 92% of individuals have access to some kind of healthcare plan and only 8% don't have an access to the healthcare plan, our dataset is imbalanced. 
We need to balance our dataset. 

## Under sampling method to balance the unbalanced data. 

```python
#### Percent Bar Plot: Healthcare Coverage---
Health_Coverage_Viz= Q1_clean["hlthpln1"].value_counts(dropna=False, normalize=True)*100
Health_Coverage_Viz


total = float(sum(Health_Coverage_Viz.values)) 
ax=sns.barplot(Health_Coverage_Viz.index, Health_Coverage_Viz.values, alpha=0.8)
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 0.3,
            '{:1.2f}'.format(height/total),
            ha="center") 
plt.title('Percentage of Survey Respondents by Health Coverage')
plt.ylabel('Percentage of Survey Respondents', fontsize=12)
plt.xlabel('Health Coverage', fontsize=12)
plt.xticks(rotation=90)
plt.show()
```

```python
#Create independent and Dependent Features
columns = Q1_clean.columns.tolist()
# Filter the columns to remove data we do not want 
columns = [c for c in columns if c not in ["hlthpln1"]]
# Store the variable we are predicting 
target = "hlthpln1"
# Define a random state 
state = np.random.RandomState(42)
X = Q1_clean[columns]
Y = Q1_clean[target]
X_outliers = state.uniform(low=0, high=1, size=(X.shape[0], X.shape[1]))
# Print the shapes of X & Y
print(X.shape)
print(Y.shape)
```

# Balancing the dataset: Under sampling Method

```python
# Import the under sampling module from imblearn library to handle the imbalance dataset
from imblearn.under_sampling import NearMiss

# Implementing Undersampling for Handling Imbalanced 
nm = NearMiss(random_state=42)
X_res,y_res=nm.fit_sample(X,Y)

X_res.shape,y_res.shape


from collections import Counter
print('Original dataset shape {}'.format(Counter(Y)))
print('Resampled dataset shape {}'.format(Counter(y_res)))
```

# Model Building using Cross Validation

```python
##### Diving the resampled data into 90% (for cross validation purposes) Vs 10% (for testing purposes)


Q1_clean_resample['is_train'] = np.random.uniform(0, 1, len(Q1_clean_resample)) <= 0.90

training_dataset, testing_dataset = Q1_clean_resample[Q1_clean_resample['is_train']==True], Q1_clean_resample[Q1_clean_resample['is_train']==False]

print("The shape of the training dataset is:",training_dataset.shape)
print("The shape of the testing dataset is:",testing_dataset.shape)
```

```python
##### Drop the is_train= True column from the training dataset and separate X and y:

training_dataset= training_dataset.drop('is_train', 1)


#### Separate the features from target variables for 90% training dataset

X_train=training_dataset[['State', 'Sex','Marital_Status','Education_Level','Housing',
                     'Veteran','Employment_Status','Children','Annual_HH_Income',
                     'Ethnicity','Race']]  # Features
y_train=training_dataset['Health_plan'] # Labels
```

```python
##### Drop the is_train= False column from the testing dataset and separate X and y:

#### Drop the is_train column that was created to identify 10% testing dataset
testing_dataset= testing_dataset.drop('is_train', 1)
testing_dataset


#### Separate the features from target variables for 90% training dataset
X_test=testing_dataset[['State', 'Sex','Marital_Status','Education_Level','Housing',
                     'Veteran','Employment_Status','Children','Annual_HH_Income',
                     'Ethnicity','Race']]  # Features
y_test=testing_dataset['Health_plan'] # Labels
```

```python
### Print the shape of X and y of training dataset(90%) and testing dataset (10%)

print("The shape of X_train (90% training dataset) is:",X_train.shape)
print("The shape of y_train (90% training dataset) is:",y_train.shape)
print("The shape of X_test (10% testing dataset) is:",X_test.shape)
print("The shape of y_test (10% testing dataset) is:",y_test.shape)
```

# Cross Validation

```python
##################  Cross Validation: for Logistic Regression only on 90 % training dataset ###############################


from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import confusion_matrix, accuracy_score


from sklearn.linear_model import LogisticRegression   #### Logistic Regression Import

#Create a Gaussian Classifier
classifier = LogisticRegression(random_state=0, solver='lbfgs', 
                         multi_class='multinomial')  ### Logistic Regression Classifier


accuracy=[]

skf=StratifiedKFold(n_splits=10, random_state=None)
skf.get_n_splits(X_train,y_train)



## X is the feature set and y is the target variable

for train_index, test_index in skf.split(X_train,y_train):
    print("Train:",train_index,"Validation:",test_index)
    X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
    y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]
    
    classifier.fit(X1_train,y1_train)
    prediction=classifier.predict(X1_test)
    score=accuracy_score(prediction, y1_test)
    accuracy.append(score)

    
### Print the accuracy of all the k fold models:
print("\n\n\n\n\nLet's print the accuracy of all the k fold models:",accuracy)



##### Let's calculate the mean accuracy of all the models of k fold validation:
print('''\n\n The mean accuracy score with 10 fold stratified CV for 
Logistic Regression algorithm on Training dataset is:''',np.array(accuracy).mean())
```
